```python
def plot_redshifts(seed=0, num=200):
    np.random.seed(seed)
    ind = np.random.randint(0, len(train_meta), num)

    specz = train_meta['hostgal_specz'].iloc[ind]
    photz = train_meta['hostgal_photoz'].iloc[ind]
    photz_err = train_meta['hostgal_photoz_err'].iloc[ind]
    fig = plt.figure(figsize=(9,7))
    ax = fig.add_subplot(111)

    ax.errorbar(specz, photz, yerr=photz_err, c='black', fmt='.')
    ax.set_xlabel('Spectroscopic Redshift')
    ax.set_ylabel('Photometric Redshift')
```

```python
plot_redshifts(0)
```

```python
plot_redshifts(0, num=400)
```
